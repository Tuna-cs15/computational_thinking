```python
#W7A1
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1
a=list(map(int,input().split()))
k=int(input())
print(binary_search(a,k))
```

```python
#W7A3
#bubble sort
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr
a=list(map(int,input().split()))
print(bubble_sort(a))
print(selection_sort(a))
```

```python
#W7A8
import ast

def bs_left(arr, L, R):
    """
    Tìm chỉ số nhỏ nhất pos sao cho:
    arr[pos][0] >= L và arr[pos][0] <= R
    Nếu không tồn tại, trả về -1
    """
    l, r = 0, len(arr) - 1
    pos = -1
    while l <= r:
        mid = (l + r) // 2
        if arr[mid][0] < L:
            l = mid + 1
        else:
            if arr[mid][0] <= R:
                pos = mid
            r = mid - 1
    return pos


s = "[[1, 4], [2, 4], [3, 6], [4, 4]]"
a = ast.literal_eval(s)

b = [2, 3, 4, 5]

for i in range(len(a)):
    a[i].append(a[i][1] - a[i][0] + 1)


a.sort(key=lambda x: (x[2], x[0], x[1]))

for i in range(len(b)):
    b[i] = [b[i], i]
b.sort(key=lambda x: x[0])

unassigned = b[:] #mảng bản sao độc lập của b

res = [-1] * len(b)

for i in a:
    L, R, length = i

    while True:
        pos = bs_left(unassigned, L, R)
        if pos == -1:
            break

        q, idx = unassigned[pos]

        res[idx] = length
        unassigned.pop(pos)   

print(res)

#Độ phức tạp: O((n+m) log m)
```

```python
#W7A9
class Node():
    def __init__(self, key, pos, left):
        self.left = None
        self.right = None
        self.val = key
        self.pos = pos 
        self.rot = left
count = 0
class BST:
    def __init__(self):
        self.root = None
        self.last = 0
    def insert(self, root, key):
        global count
        if root is None:
            node = Node(key, count, 0)
            self.last = node.pos
            count = 0
            return node
        else:
            root.pos = max(count+root.rot, root.pos)
            if root.val < key:
                count = root.pos + 1
                root.right = self.insert(root.right, key)
            else:
                root.pos += 1
                root.rot += 1
                root.left = self.insert(root.left, key)
        return root
a=list(map(int,input().split()))
a=a[::-1]
tree=BST()
root = None
res=[]
for i in a:
    root=tree.insert(root, i)
    res.append(tree.last)
res=res[::-1]
print(res)
```

```python
#W7A10
import bisect

def max_subarray_mod(nums, m):
    prefix = 0
    best = 0
    prefixes = []

    for x in nums:
        prefix = (prefix + x) % m
        best = max(best, prefix)

        i = bisect.bisect_right(prefixes, prefix)
        if i < len(prefixes):
            best = max(best, (prefix - prefixes[i] + m) % m)

        bisect.insort(prefixes, prefix)

    return best


# TEST
print(max_subarray_mod([3, 3, 9, 9, 5], 7))     # 6
print(max_subarray_mod([1, 2, 3], 2))          # 1
print(max_subarray_mod([10, 20, 30], 15))      # 10
print(max_subarray_mod([7,11,14,2,8,4], 12))   # 11
print(max_subarray_mod([1,5,9,13,17], 10))     # 9
```

```python
#W7A11
def count_word(sentence, keyword):
    words = sentence.split()
    count = 0
    for w in words:
        if w == keyword:
            count += 1
    return count


# TEST
print(count_word("There's a fire that burns inside.", "fire"))  # 1
print(count_word("die die die", "die"))                         # 3
```

```python
#W7A12
def fastest_finger(words):
    labels = ['A', 'B', 'C', 'D']
    pairs = []

    for i in range(4):
        pairs.append((words[i], labels[i]))

    pairs.sort()   # sắp xếp theo từ điển
    result = []

    for _, label in pairs:
        result.append(label)

    return " ".join(result)


# TEST
print(fastest_finger(["May", "Troi", "Cua", "Em"]))  
print(fastest_finger(["Hoa hong", "hoa hue", "hoa lan", "hoa huong duong"]))
```

```python
#W7A13
from collections import Counter

def longest_harmonious_subsequence(arr):
    freq = Counter(arr)
    best = 0

    for x in freq:
        if x + 1 in freq:
            best = max(best, freq[x] + freq[x + 1])

    return best


# TEST
print(longest_harmonious_subsequence([1,3,2,2,5,2,3,7]))  # 5
```

```python
#W7A14
def find_all_sevens(arr):
    positions = []

    for i in range(len(arr)):
        if arr[i] == 7:
            positions.append(i)

    if not positions:
        return "Not found"

    positions.sort(reverse=True)
    return " ".join(map(str, positions))


# TEST
print(find_all_sevens([7, 10, -2, 12, -1]))  # 0
print(find_all_sevens([1, 2, 3]))            # Not found
```
